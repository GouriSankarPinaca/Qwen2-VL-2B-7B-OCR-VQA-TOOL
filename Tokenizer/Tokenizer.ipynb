```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

#example text
text = "let's figer it out"

# tokenizer the text
token = tokenizer.tokenize(text)
print(token)

# convert token to input IDs
input_ids = tokenizer.convert_tokens_to_ids(token)
print(input_ids)

# Encode the text
enumerated_text = tokenizer.encode(text)
print(enumerated_text)

# Decode the text
text2 = tokenizer.decode(input_ids)
print(text2)

```
